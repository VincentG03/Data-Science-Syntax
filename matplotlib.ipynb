# Plotting with Matplotlib

Matplotlib is a powerful plotting library in Python. Here are examples of how to create different types of plots using `matplotlib.pyplot`, with all datasets defined at the start.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Displays the graph under the code instead of creating a pop up
%matplotlib inline
```

---
# Scatterplot

A scatterplot is used to visualize the relationship between two numerical variables. Each point on the plot represents an observation in the dataset, with its position determined by the values of the two variables.

### When to Use?
- **Purpose**: To investigate potential relationships, correlations, or patterns between two numerical variables.


```python
# Scatter Plot Data
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [10, 15, 13, 17, 20]

# Scatter Plot
plt.scatter(scatter_x, scatter_y, color='blue', marker='o')
plt.title('Scatter Plot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
```

---
# Histogram

A histogram displays the distribution of a dataset by grouping data points into bins and showing the frequency of data points within each bin.

### When to Use?
- **Purpose**: To understand the distribution and frequency of a numerical variable.


```python
# Histogram Data
hist_data = [10, 15, 13, 17, 20, 15, 10, 13, 18, 17, 20, 22]

# Histogram
plt.hist(hist_data, bins=5, color='lightblue', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

---
# Barplot

A bar plot represents categorical data with rectangular bars, where the length of each bar is proportional to the value it represents.

### When to Use?
- **Purpose**: To understand the distribution and frequency of a categorical variable.


```python
# Bar Plot Data
bar_categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 8]

# Bar Plot
plt.bar(bar_categories, bar_values, color='lightblue')
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

---
# Pie Chart

A pie chart shows proportions of a whole as slices of a circle, where each slice represents a category’s contribution to the total.

### When to Use?
- **Purpose**: To show relative proportions or percentages of a categorical variable.


```python
# Pie Chart Data
pie_sizes = [15, 30, 45, 10]
pie_labels = ['A', 'B', 'C', 'D']
pie_colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

# Pie Chart
plt.pie(pie_sizes, labels=pie_labels, colors=pie_colors, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Example')
plt.show()
```

---
# Line Chart 

A line plot is used to show trends over time or ordered categories by connecting data points with a continuous line.

### When to Use?
- **Purpose**: To display trends or changes in data over a continuous range or time (numerical and categorical).


```python
# Line Plot Data
line_x = [1, 2, 3, 4, 5]
line_y = [10, 15, 13, 17, 20]

# Line Plot
plt.plot(line_x, line_y, color='darkblue', marker='o')
plt.title('Line Chart Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
```

---
# Box Plot

A box plot displays the distribution of a dataset based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It is useful for understanding the spread and identifying outliers in the data.

### When to Use?
- **Purpose**: To show the distribution, central tendency, and variability of a dataset, as well as to identify outliers.

```python
# Box Plot Data
box_data = [10, 15, 13, 17, 20, 15, 10, 13, 18, 17, 20, 22]

# Box Plot (box_data, vertical, colour_in, colour)
plt.boxplot(box_data, vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Box Plot Example')
plt.show()
```

# Heatmap

A heatmap visualizes matrix-like data with color-coding, where each cell's color represents the value of the corresponding data point. It is useful for identifying patterns and intensities across two dimensions.

### When to Use?
- **Purpose**: To visualize the intensity or magnitude of values in a matrix and identify patterns or correlations.

```python
"""
For "interpoolation" try: 
'nearest': Fast, blocky appearance.
'bilinear': Smoother transitions, moderate speed.
'bicubic': Very smooth, computationally intensive.
'lanczos': High-quality, very slow for large datasets.
"""

heatmap_data = pd.DataFrame([
    [1.0, 0.8, 0.6, 0.2],
    [0.8, 1.0, 0.5, 0.1],
    [0.6, 0.5, 1.0, 0.3],
    [0.2, 0.1, 0.3, 1.0]
])

# Heatmap (data, colour_scheme, how to blend colours)
plt.imshow(heatmap_data, cmap='hot', interpolation='nearest') 
plt.title('Heatmap Example')
plt.colorbar()  # Show color scale
plt.show()
```
